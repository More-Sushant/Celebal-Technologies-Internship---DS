```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Loading the dataset
heart_df = pd.read_csv("heart.csv")
heart_df.head()
```

```python
# Line plot
chol_by_age = heart_df.groupby("age")["chol"].mean()

plt.figure(figsize=(10,6))
plt.plot(chol_by_age.index, chol_by_age.values, marker='o')
plt.title("Average Cholesterol by Age")
plt.xlabel("Age")
plt.ylabel("Average Cholesterol")
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.hist(heart_df['thalach'], bins=20, color='skyblue', edgecolor='black')
plt.title("Distribution of Max Heart Rate Achieved")
plt.xlabel("Max Heart Rate")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

```python
sex_counts = heart_df['sex'].value_counts()
labels = ['Male', 'Female']
plt.pie(sex_counts, labels=labels, autopct='%1.1f%%', colors=['#66b3ff', '#ff9999'])
plt.title("Sex Distribution in Dataset")
plt.show()
```

```python

plt.figure(figsize=(8,6))
sns.boxplot(x='target', y='chol', data=heart_df, palette='pastel')
plt.title("Cholesterol Levels vs Heart Disease Presence")
plt.xlabel("Heart Disease (1 = Yes, 0 = No)")
plt.ylabel("Cholesterol")
plt.grid(True)
plt.show()
```
